```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
measurements = pd.read_csv('https://raw.githubusercontent.com/samielzaret7/technical_challenge_DA/master/data/measurements_clean.csv',index_col=0)
df_e10 = pd.read_csv('https://raw.githubusercontent.com/samielzaret7/technical_challenge_DA/master/data/e10.csv',index_col=0)
df_sp98 = pd.read_csv('https://raw.githubusercontent.com/samielzaret7/technical_challenge_DA/master/data/sp98.csv',index_col=0)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(df_e10['distance'], df_e10['consume'], label='E10', color='blue', alpha=0.7)

plt.scatter(df_sp98['distance'], df_sp98['consume'], label='SP98', color='red', alpha=0.7)

plt.xlabel('Distance (km)')
plt.ylabel('Fuel Consumption (L/100km)')
plt.title('Distance vs Fuel Consumption')

plt.legend()

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(df_e10['Fuel Consumption'], df_e10['Price of Consumption'], label='E10', color='blue', alpha=0.7)

plt.scatter(df_sp98['Fuel Consumption'], df_sp98['Price of Consumption'], label='SP98', color='red', alpha=0.7)

plt.xlabel('Fuel Consumption (liters)')
plt.ylabel('Cost of Consumption')
plt.title('Fuel Consumption vs Cost of Consumption')

plt.legend()

plt.grid(True)
plt.show()
```
